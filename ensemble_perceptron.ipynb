```python
import pandas as pd
import torch, numpy

device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
```

```python
train_data = pd.read_csv('train_outputs.csv')
validation_data = pd.read_csv('validation_outputs.csv')
evaluation_data = pd.read_csv('evaluation_outputs.csv')
```

```python
class EnsemblePerceptron(torch.nn.Module):
    def __init__(self):
        super(EnsemblePerceptron, self).__init__()
        self.fc = torch.nn.Linear(2,1)
        self.sigmoid = torch.nn.Sigmoid()
    def forward(self, x):
        output = self.fc(x)
        output = self.sigmoid(output) 
        return output
```

```python
X_train = pd.DataFrame(train_data[['Predicted_Label_Mouth','Predicted_Label_Eyes']]).to_numpy()
X_train = torch.FloatTensor(X_train).to(device)

Y_train = pd.DataFrame(train_data['True Label']).to_numpy()
Y_train = torch.FloatTensor(Y_train).to(device)


X_validation = pd.DataFrame(validation_data[['Predicted_Label_Mouth','Predicted_Label_Eyes']]).to_numpy()
X_validation = torch.FloatTensor(X_validation).to(device)

Y_validation = pd.DataFrame(validation_data['True Label']).to_numpy()
Y_validation = torch.FloatTensor(Y_validation).to(device)


X_evaluation = pd.DataFrame(evaluation_data[['Predicted_Label_Mouth','Predicted_Label_Eyes']]).to_numpy()
X_evaluation = torch.FloatTensor(X_evaluation).to(device)

Y_evaluation = pd.DataFrame(evaluation_data['True Label']).to_numpy()
Y_evaluation= torch.FloatTensor(Y_evaluation).to(device)
```

```python
torch.manual_seed(720)
model = EnsemblePerceptron()
model = model.to(device)
```

```python
for param in model.parameters():
    print('Weights:' + str(param.data))
```

```python
criterion = torch.nn.BCELoss()
optimizer = torch.optim.SGD(model.parameters(), lr = 0.01)
```

```python
for epoch in range(50000):
    model.train()
    optimizer.zero_grad()
    # Forward pass
    y_pred = model(X_train)
    # Compute Loss
    loss = criterion(y_pred, Y_train)
    if epoch % 100 == 0:
        print(loss)
        #for param in model.parameters():
        #    print('Weights:' + str(param.data))
    # Backward pass
    loss.backward()
    optimizer.step()
```

```python
model.eval()
model=model.to(device)
```

```python
y_pred_train = model(X_train.to(device))
y_pred_validation= model(X_validation.to(device))
y_pred_evaluation= model(X_evaluation.to(device))
```

```python
from sklearn.metrics import classification_report
```

```python
m = torch.nn.Threshold(0.5, 1)
y_pred_train= torch.round(torch.FloatTensor(y_pred_train.to(torch.device("cpu")).detach().numpy()))
y_pred_validation= torch.round(torch.FloatTensor(y_pred_validation.to(torch.device("cpu")).detach().numpy()))
y_pred_evaluation= torch.round(torch.FloatTensor(y_pred_evaluation.to(torch.device("cpu")).detach().numpy()))
```
